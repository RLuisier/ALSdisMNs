```python
%%R

#
#NESTED FUNCTIONS
#

                  
testLogNorm <- function(val){
  if(min(val)<0){
    return(FALSE)
  }
  if(min(val)>=0){
    val  <- val+min(val[val>0])
    valt <- log(val)
    set.seed(10)
    m1   <- lm(quantile(val,probs=seq(0, 1, 0.01))~quantile(rnorm(n=length(val),mean(val),sd(val)),probs=seq(0, 1, 0.01)))
    set.seed(10)
    m2   <- lm(quantile(valt,probs=seq(0, 1, 0.01))~quantile(rnorm(n=length(valt),mean(valt),sd(valt)),probs=seq(0, 1, 0.01)))
    return(-anova(m1)$F[1]+anova(m2)$F[1]>10)
  }
}                     
                   
                                
                                 
ModifyImageID <- function(idtm=MNs$Metadata_ID){
    idtm      <- gsub(as.character(idtm),pattern="-CHAT.tif",repl="")
    idtm      <- gsub(as.character(idtm),pattern="2ki",repl="4Ki")
    idtm      <- gsub(as.character(idtm),pattern="KI",repl="Ki")
    idtm      <- gsub(as.character(idtm),pattern="ki",repl="Ki")
    idtm      <- gsub(as.character(idtm),pattern="^TG",repl="1TG")
    idtm      <- gsub(as.character(idtm),pattern="ki",repl="Ki")
    return(idtm)
}
                                 
                               
```

```python
def GMM_analysis_human_mv(myvars,dfc,limangle,fst,batch,AB,title,colCT,colALS,Plot):
    
    random_state = np.random.RandomState(seed=1)
    dfc['group']=dfc['mutation']+'-'+dfc['ID'];
    datoi=(dfc[myvars],
                   dfc.groupby('mutation')[myvars].get_group('CTRL'),
                   dfc.groupby('mutation')[myvars].get_group('sALS'));

    myX=dfc[myvars].to_numpy()
    gm = GaussianMixture(n_components=2,  n_init=20,covariance_type='full',init_params='kmeans')
    gm.fit(myX)


    dfc['p1']=gm.predict_proba(myX)[:,0]
    if np.argmax(dfc.groupby(['mutation'])['p1'].mean())=='sALS':
        mu_idx=0;
        ctrl_idx=1;
    else:
        mu_idx=1;
        ctrl_idx=0;

    dfc['posterior_probability']=gm.predict_proba(myX)[:,ctrl_idx];
    dfc['severity']=-np.log10(gm.predict_proba(myX)[:,ctrl_idx]);
    dft=dfc.sort_values(by='severity',ascending=True)
    dft=dft.loc[np.diff(dft['severity'])>0]

    xs=np.array(dft['severity'])
    mymax=np.max(xs)
    xs=xs/mymax
    ys=np.array(dft['posterior_probability'])
    ysg = gaussian_filter(ys, sigma=0.1)
    angles=np.degrees(np.arctan(ysg/xs))
    idx=np.argmax((angles<=limangle) & (angles>0))
    xs=xs*mymax
    severity_lim=dft.iloc[idx]['severity'];
    lowerbound=dft.iloc[idx]['posterior_probability'];
    limpc5=dft.iloc[idx][myvars];

    dfc['test']=dfc['posterior_probability']<lowerbound
    d = {'fraction':100*dfc.groupby(['ID'])['test'].apply(sum)/dfc.groupby(['ID'])['test'].apply(len),
         'avg_prob':1-dfc.groupby(['ID'])['posterior_probability'].mean()}

    summary_pc = pd.DataFrame(d)
    summary_pc['ntot']=0
    summary_pc['nsick']=0
    summary_pc['severity_global']=0
    summary_pc['severity_focus']=0
    summary_pc['ncr_global']=0
    summary_pc['ncr_focus']=0
    summary_pc['htest']=0
    summary_pc['Shannon']=0
    summary_pc['probability_sick']=0
    summary_pc['like_sick']=0
    mytemp = dfc.drop_duplicates('ID')
    mytemp.index=mytemp['ID']
    summary_pc['mutation']=mytemp['mutation']



    for myid in dfc['ID'].unique():
            summary_pc.loc[myid,'ntot']=dfc.loc[(dfc['ID']==myid)].shape[0]
            summary_pc.loc[myid,'nsick']=dfc.loc[(dfc['ID']==myid)&(dfc['posterior_probability']<lowerbound)].shape[0]
            summary_pc.loc[myid,'severity_focus']=dfc.loc[(dfc['ID']==myid)&(dfc['posterior_probability']<lowerbound),
                                                          'severity'].mean()
            summary_pc.loc[myid,'severity_global']=dfc.loc[dfc['ID']==myid,'severity'].mean()

            summary_pc.loc[myid,'Shannon']=(-sum(dfc.loc[dfc['ID']==myid,'posterior_probability']*np.log(dfc.loc[dfc['ID']==myid,'posterior_probability'])))

            var=multivariate_normal(gm.means_[mu_idx], cov=gm.covariances_[mu_idx])
            like_mu=var.pdf(dfc.loc[dfc['ID']==myid,myvars]).prod()
            var=multivariate_normal(gm.means_[ctrl_idx], cov=gm.covariances_[ctrl_idx])
            like_ctrl=var.pdf(dfc.loc[dfc['ID']==myid,myvars]).prod()

            summary_pc.loc[myid,'probability_sick']=like_mu/(like_mu+like_ctrl)
            summary_pc.loc[myid,'like_sick']=np.log10(like_mu)

            M=summary_pc.loc[summary_pc['mutation']=='CTRL','ntot'].sum()+summary_pc.loc[myid,'ntot']
            n=summary_pc.loc[summary_pc['mutation']=='CTRL','nsick'].sum()+summary_pc.loc[myid,'nsick']
            N=summary_pc.loc[myid,'ntot']
            X=summary_pc.loc[myid,'nsick']-1
            summary_pc.loc[myid,'htest']=hypergeom.sf(X, M, n, N)

    summary_pc.sort_values(by=['mutation'],inplace=True)

    fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(16, 8))
    st = fig.suptitle(title, fontsize="x-large")
    plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.3, hspace=0.5)

    plt.subplot(2,4,1)
    x1 = np.linspace(myX.min(axis=0)[0], myX.max(axis=0)[0], 1000)
    x2 = np.linspace(myX.min(axis=0)[1], myX.max(axis=0)[1], 1000)
    X1,X2 = np.meshgrid(x1, x2)
    XX = np.array([X1.ravel(), X2.ravel()]).T
    logprob = gm.score_samples(XX)
    responsibilities = gm.predict_proba(XX)
    pdf = np.exp(logprob)
    pdf_individual = responsibilities * pdf[:, np.newaxis]
    Z = -gm.score_samples(XX)
    Z = Z.reshape(X1.shape)
    CS = plt.contour(X1, X2, Z, norm=LogNorm(vmin=0.1, vmax=100.0),levels=np.logspace(-100, 100, 2500),alph=0.5)
    #CB = plt.colorbar(CS, shrink=0.8, extend='both')
    plt.scatter(myX[dfc['mutation']=='sALS', 0], myX[dfc['mutation']=='sALS', 1], 20.0,color=colALS)
    plt.scatter(myX[dfc['mutation']=='CTRL', 0], myX[dfc['mutation']=='CTRL', 1], 20.0,color=colCT)
    plt.xlabel(myvars[0],fontsize=12)
    plt.ylabel(myvars[1],fontsize=12)
    #plt.axvline(x=limpc5[0],ls='--', color='maroon',linewidth=1.5)
    #plt.axhline(y=limpc5[1],ls='--', color='maroon',linewidth=1.5)

    plt.subplot(2,4,2)
    temp=pd.DataFrame({'name':dfc['group'].unique()})
    temp['color']=colCT
    temp.loc[['sALS' in i for i in temp['name']],'color']=colALS
    colors=temp['color']
    bplot1 = sns.boxplot(y='posterior_probability', x='group', data=dfc, width=0.5,palette=colors,showfliers=False)
    bplot1 = sns.swarmplot(y='posterior_probability', x='group', data=dfc, color="black",  size=3)
    plt.grid(True,color='grey', linestyle=':', linewidth=1)
    plt.ylabel('Posterior probability P\n to be healthy',fontsize=12)
    plt.xticks(fontsize=fst,rotation=45)
    plt.xlabel('')
    plt.axhline(y=lowerbound,ls='--', color='maroon',linewidth=1.5)

    plt.subplot(2,4,3)   
    frac=100*dfc.groupby(['group'])['test'].apply(sum)/dfc.groupby(['group'])['test'].apply(len)
    temp=pd.DataFrame(frac)
    temp['color']=colCT
    temp.loc[['sALS' in i for i in temp.index],'color']=colALS
    colors=temp['color']
    x=np.arange(1,len(frac)+1)
    plt.barh(x,frac,color=colors)
    plt.xlim(0,100)
    plt.grid()
    plt.xlabel('% of sick cells',fontsize=12)
    plt.yticks([r + 1 for r in range(len(frac))], pd.DataFrame(frac).index,fontsize=fst)
    temp['mutation']='CTRL'
    temp.loc[['sALS' in i for i in temp.index],'mutation']='sALS'
    plt.ylabel('')


    plt.subplot(2,4,4)   
    colors=[colCT,colALS]
    bplot1 = sns.boxplot(y='test', x='mutation', data=temp, width=0.5,palette=colors,showfliers=False)
    bplot1 = sns.swarmplot(y='test', x='mutation', data=temp, color="black",  size=8)
    plt.grid(True,color='grey', linestyle=':', linewidth=1)
    plt.ylabel('% predicted sick cells',fontsize=12)
    tempval=ttest_ind(a=temp.groupby('mutation')['test'].get_group('CTRL'),
                             b=temp.groupby('mutation')['test'].get_group('sALS'),equal_var=False).pvalue/2
    plt.title( "P-value(Welsch test)="+str('%.2E' % tempval))
    plt.xticks(fontsize=12)
    plt.xlabel('')


    colors = [colCT,colALS]
    names = ['CTRL', 'sALS']

    ixp=5
    for n in [0,1] :

        plt.subplot(2,4,ixp)

        datoi=(dfc[myvars[n]],
           dfc.groupby('mutation')[myvars[n]].get_group('CTRL'),
           dfc.groupby('mutation')[myvars[n]].get_group('sALS'));
        plt.hist([datoi[1], datoi[2]], bins = 25, normed=True,color = colors, label=names,alpha=0.5);

        x = np.linspace(np.min(datoi[0]), np.max(datoi[0]), 1000)
        density = gaussian_kde(datoi[0])
        density.covariance_factor = lambda : .25
        density._compute_covariance()
        plt.plot(x,density(x),'k', color = 'black',linewidth=2.5,label='data')
        colors = ['red', 'green'];
        for i, color in enumerate(colors, start=0):
                plt.plot(x, norm.pdf(x=x.reshape(-1, 1), loc=gm.means_[i][n], 
                                scale=np.sqrt(gm.covariances_[n][i,i])),'--k', color=color,linewidth=2.0)
        plt.grid(color='grey', linestyle=':', linewidth=1)
        plt.title( "Gaussian mixture modeling")
        plt.xlabel(myvars[n],fontsize=12)
        plt.ylabel('$p(x)$')
        plt.legend(loc='best')
        plt.axvline(x=limpc5[n],ls='--', color='maroon',linewidth=1.5)

        colors = [colCT,colALS]
        names = ['CTRL', 'sALS']

        ixp=ixp+1
        plt.subplot(2,4,ixp)

        temp=pd.DataFrame({'name':dfc['group'].unique()})
        temp['color']=colCT
        temp.loc[['sALS' in i for i in temp['name']],'color']=colALS
        colors=temp['color']
        bplot1 = sns.boxplot(y=myvars[n], x='group', data=dfc, width=0.5,palette=colors,showfliers=False)
        bplot1 = sns.swarmplot(y=myvars[n], x='group', data=dfc, color="black",  size=3)
        plt.grid(True,color='grey', linestyle=':', linewidth=1)
        plt.ylabel(myvars[n],fontsize=12)
        plt.xlabel('')
        plt.xticks(fontsize=fst,rotation=45)
        plt.axhline(y=limpc5[n],ls='--', color='maroon',linewidth=1.5)
        ixp=ixp+1
        colors = [colCT,colALS]
        names = ['CTRL', 'sALS']
                                  
    plt.subplots_adjust(hspace=0.5,wspace=0.5)
                                  
    return dfc,limpc5,summary_pc,dfc['test'],dfc['posterior_probability']
```

```python
%%R -o Us_sfpq -o subdat_sfpq -o subinfo_sfpq -o mat_sfpq

AB="SFPQ"
subdat_sfpq                 <- data[myinfo$rbp1==AB,]
subinfo_sfpq                <- myinfo[myinfo$rbp1==AB,]
colnames(subdat_sfpq)       <- gsub(colnames(subdat_sfpq),pattern="RBP1",repl=AB)
subinfo_sfpq$patient        <- factor(as.character(subinfo_sfpq$patient))
subinfo_sfpq$mutation       <- factor(as.character(subinfo_sfpq$Group))
subinfo_sfpq$ID             <- factor(subinfo_sfpq$ID)


#Remove measurements with NAs -same on full and sel
subdat_sfpq                        <- subdat_sfpq[,apply(is.na(subdat_sfpq),2,sum)==0]
mat_sfpq                           <- data.matrix(subdat_sfpq)

#Test for log-normality of the data
dF               <- unlist(lapply(c(1:ncol(mat_sfpq)),function(ix)return(testLogNorm(mat_sfpq[,ix]))))&apply(mat_sfpq,2,min)>=0
for(ix in which(dF)){
  mat_sfpq[,ix]<-log(mat_sfpq[,ix]+min(mat_sfpq[mat_sfpq[,ix]>0,ix]))
}

#Standardized the measurements across the MNs
myavg <- apply(mat_sfpq,2,function(Z)return(mean(Z,na.rm=T)))  
mysd  <- apply(mat_sfpq,2,function(Z)return(sd(Z,na.rm=T)))  
for(i in c(1:ncol(mat_sfpq))){
    mat_sfpq[,i]     <- (mat_sfpq[,i]-myavg[i])/mysd[i]
}
               
mat_sfpq              <- t(mat_sfpq)             
mat_sfpq              <- mat_sfpq[apply(is.na(mat_sfpq),1,sum)==0,]
     
#SVD
SVD_eset    <- svd(mat_sfpq)
pi          <- SVD_eset$d*SVD_eset$d/sum(SVD_eset$d*SVD_eset$d)
d           <- -1*sum(pi*log10(pi))/log10(length(pi))
Us          <- data.frame(SVD_eset$u)
rownames(Us)<- rownames(mat_sfpq)
colnames(Us)<- paste("PC",c(1:ncol(Us)),sep="")
ncomp        <- which(cumsum(pi)>=0.9)[1]

#LMM to test for association between ALS and each component                                  
test_ALS_SFPQ <- do.call(what=rbind,args=lapply(c(1:ncomp),function(IX){
  tempdat                <- data.frame(PC=SVD_eset$v[,IX],subinfo_sfpq)
  m1.full                <- lmer(PC~1+mutation+(1|patient), data = tempdat)
  m1.null                <- lmer(PC~1+(1|patient), data = tempdat)
  tempdat=data.frame(anova(m1.null,m1.full))
  tempdat2=data.frame(coef(summary(m1.full)))
  return(c(tempdat['m1.full','Chisq'],tempdat['m1.full','Pr..Chisq.'],tempdat2['mutationsALS','Estimate'],tempdat2['mutationsALS','Std..Error']))
}))
                                  
colnames(test_ALS_SFPQ)<-c("Chisquare","Pvalue","effect_avg","effect_sd")
rownames(test_ALS_SFPQ)<-paste("PC",c(1:ncomp),sep="")
test_ALS_SFPQ <- data.frame(test_ALS_SFPQ)
sel_ALS_SFPQ  <- c(1:ncomp)[test_ALS_SFPQ$Pvalue<=0.05]

                                  
#Because data have been scaled projection and correlation are equivalent
myCorr <- do.call(what=cbind,args=lapply(c(1:ncomp),function(IX){
  return(unlist(lapply(c(1:nrow(mat_sfpq)),function(ix)return(cor(x=SVD_eset$v[,IX],y=mat_sfpq[ix,],method='pearson')))))
}))
rownames(myCorr)<- rownames(mat_sfpq)
         
temp           <- SVD_eset$v[,c(1:ncomp)]
colnames(temp) <- paste("PC",c(1:ncomp),sep="")        
subinfo_sfpq   <- data.frame(subinfo_sfpq,temp)
Us_sfpq        <- Us         
         
```

```python
%%R -i subdat_sfpq -i subinfo_sfpq -i subdat_fus_full -i subinfo_fus_full


TestNC_localisation <- function(myvar,LOG=FALSE,subdat=subdat_sfpq,subinfo=subinfo_sfpq,testCol='test_LR'){
  par(mar=c(3,4,5,1))
  if(LOG){
     tempdat                <- data.frame(PC=log(subdat[,myvar]),subinfo)
  } else {
     tempdat                <- data.frame(PC=subdat[,myvar],subinfo) 
  }
    
  tempdat$mutant         <- factor(ifelse(tempdat[,testCol],'sALS','CTRL'),levels=c("CTRL",'sALS'))
  m1.full                <- lmer(PC~1+mutant+(1|patient), data = tempdat)
  m1.null                <- lmer(PC~1+(1|patient), data = tempdat)
  tempdat3=data.frame(anova(m1.null,m1.full))
  tempdat2=data.frame(coef(summary(m1.full)))
  var_tests=data.frame(c(tempdat3['m1.full','Chisq'],tempdat3['m1.full','Pr..Chisq.'],
                         tempdat2['mutantsALS','Estimate'],tempdat2['mutantALS','Std..Error']))
  rownames(var_tests)<-c("Chisquare","Pvalue","effect_avg","effect_sd")
    
  tempCols        <- c("#928E8D","#81BA83")
  names(tempCols) <- c("CTRL","sALS")
  myCols          <- tempCols[match(subinfo$mutation[match(levels(factor(tempdat$IDmore)),subinfo$IDmore)],names(tempCols))]

  mymethod='hex'  
  boxplot(PC~mutant,data=tempdat,outline=FALSE,las=1,xlab="",ylab=myvar,col=c("white"),frame=F,main="")
  beeswarm(PC~mutant,data=tempdat[tempdat$mutation=='CTRL',],bg="#928E8D",col=NA,add=TRUE,pch=21,cex=1.8,method=mymethod)
  beeswarm(PC~mutant,data=tempdat[tempdat$mutation=='sALS',],bg="#81BA83",col=NA,add=TRUE,pch=21,cex=1.8,method=mymethod) 
  voi1=formatC(var_tests["Chisquare",1],format="e",digit=1)
  voi2=formatC(var_tests["Pvalue",1],format="e",digit=2)
  voi3=formatC(var_tests["effect_avg",1],digit=2)
  voi4=formatC(var_tests["effect_sd",1],digit=2)
  mtext(side=3,line=1,text=expr(paste("ALS effect by about ",!!voi3, "\u00B1",!!voi4," (sd) ",sep="")),cex=0.9)
  mtext(side=3,line=0,text=expr(paste(chi^2,"(1) = ",!!voi1,", P = ",!!voi2,sep="")),cex=0.9)
  mtext(side=3,line=3,text=testCol,cex=0.9,font=2)
  grid()  
    
}
```
